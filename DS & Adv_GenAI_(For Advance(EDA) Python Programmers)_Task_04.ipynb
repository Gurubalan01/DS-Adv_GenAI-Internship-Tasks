```python
class Solution(object):
    def maxProduct(self, nums):
        nums.sort()                
        return (nums[-1] - 1) * (nums[-2] - 1)
obj = Solution()
print(obj.maxProduct([3,4,5,2]))
```

```python
class Solution(object):
    def numTeams(self, rating):
        n = len(rating)
        teams = 0

        for j in range(n):
            left_smaller = left_greater = 0
            right_smaller = right_greater = 0


            for i in range(j):
                if rating[i] < rating[j]:
                    left_smaller += 1
                else:
                    left_greater += 1


            for k in range(j + 1, n):
                if rating[k] > rating[j]:
                    right_greater += 1
                else:
                    right_smaller += 1

            teams += (left_smaller * right_greater)
            teams += (left_greater * right_smaller)

        return teams
obj = Solution()
print(obj.numTeams([2,5,3,4,1]))
```

```python
class Solution(object):
    def busyStudent(self, startTime, endTime, queryTime):
        count = 0

        for i in range(len(startTime)):
            if startTime[i] <= queryTime <= endTime[i]:
                count += 1

        return count
obj = Solution()
print(obj.busyStudent([1,2,3],[3,2,7],4))
```

```python
class Solution(object):
    def numberOfSteps(self, num):
        steps = 0

        while num > 0:
            if num % 2 == 0:
                num //= 2
            else:
                num -= 1
            steps += 1

        return steps
obj = Solution()
print(obj.numberOfSteps(14))
```

```python
class Solution(object):
    def countBits(self, n):
        ans = [0] * (n + 1)

        for i in range(1, n + 1):
            ans[i] = ans[i >> 1] + (i & 1)

        return ans
obj = Solution()
print(obj.countBits(2))
```
